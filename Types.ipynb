```python
import pandas as pd
#df = pd.read_csv("cancer.csv") #reads in csv file
df.types = pd.read_csv("cancer.csv", usecols = ["State","Total.Rate","Types.Breast.Total", "Types.Colorectal.Total", "Types.Lung.Total"])
df.types.drop(4) #drop California
```

```python
#correlation matrix
df.types.corr()
```

```python
import matplotlib.pyplot as plt

print("Types.Breast.Total vs. Total.Rate")
plt.scatter(df["Types.Breast.Total"], df["Total.Rate"]) #scatter
plt.show()

print("Types.Colorectal.Total vs. Total.Rate")
plt.scatter(df["Types.Colorectal.Total"], df["Total.Rate"]) #scatter
plt.show()

print("Types.Lung.Total vs. Total.Rate")
plt.scatter(df["Types.Lung.Total"], df["Total.Rate"]) #scatter
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error
print("Linear Regression for Types.Lung.Total and Total Rate:")
x = np.array(df["Types.Lung.Total"]).reshape(-1,1)
y = np.array(df["Total.Rate"]) #y is the same for each of these linear regression models
model1 = LinearRegression(fit_intercept=True, normalize=False)
model1.fit(x,y)
#predict new data
y_predict = model1.predict(x)
#slope and intercept and error
print("Slope:", model1.coef_, ", Intercept:", model1.intercept_, ", Error:", metrics.mean_squared_error(y, y_predict))
#step 7, line of best fit
plt.scatter(x,y_predict)
plt.xlabel("Types.Lung.Total")
plt.ylabel("Total.Rate")
plt.show()
```
